```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

train_dir = pd.read_csv('train.csv')
test_dir = pd.read_csv('test.csv')
print("Training data: {}".format(train_dir.shape))
print("Testing data: {}".format(test_dir.shape))
```

```python
train_dir.head()
```

```python
gender = {
    '?': 0,
    'Pekerja Bebas Bukan Perusahan': 1,
    'Pekerja Bebas Perusahaan': 2,
    'Pemerintah Lokal': 3,
    'Pemerintah Negara': 4,
    'Pemerintah Provinsi': 5,
    'Tanpa di Bayar': 6,
    'Tidak Pernah Bekerja': 7,
    'Wiraswasta': 8,
}
train_dir['Kelas Pekerja'] = train_dir['Kelas Pekerja'].apply(lambda x: gender.get(x))
test_dir['Kelas Pekerja'] = test_dir['Kelas Pekerja'].apply(lambda x: gender.get(x))

train_dir.head()
```

```python
updated_train_dir = train_dir.drop(['id', 'Umur','Jmlh Tahun Pendidikan',
                                   'Jenis Kelamin','Keuntungan Kapital', 'Kerugian Capital','Jam per Minggu'], axis = 1)
updated_train_dir.head(3)
```

```python
updated_test_dir = test_dir.drop(['id', 'Umur','Jmlh Tahun Pendidikan',
                                   'Jenis Kelamin','Keuntungan Kapital', 'Kerugian Capital','Jam per Minggu'], axis = 1)
updated_test_dir.head(3)
```

```python
gender = {
    '>7jt': 1,
    '<=7jt': 0
}
updated_train_dir['Gaji'] = updated_train_dir['Gaji'].apply(lambda x: gender.get(x))

updated_train_dir.head()
```

```python
traindf = pd.get_dummies(updated_train_dir, columns = ["Kelas Pekerja","Pendidikan","Status Perkawinan", "Pekerjaan"],
                             prefix=["kelas_","Pd_", "Status_", "kerja_"])
testdf = pd.get_dummies(updated_test_dir, columns = ["Kelas Pekerja","Pendidikan","Status Perkawinan", "Pekerjaan"],
                             prefix=["kelas_","Pd_", "Status_", "kerja_"])
traindf.head()
```

```python
traindf.head()
```

```python
train_y = traindf['Gaji']
traindf.drop(['Gaji'], axis=1,inplace=True)
traindf.head()
```

```python
# Split data
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(traindf,
                                                   train_y,
                                                test_size=0.3,
                                                   random_state=42)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()

min_max_scaler.fit(X_train)

X_train_scaled = min_max_scaler.transform(X_train)

X_test_scaled = min_max_scaler.transform(X_test)

testdf_scaled = min_max_scaler.transform(testdf)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.linear_model import LogisticRegression
LogisticRegression = LogisticRegression(max_iter=10000)
LogisticRegression.fit(X_train_scaled, Y_train)
```

```python
predictions = LogisticRegression.predict(X_test_scaled)
predictions
```

```python
Linear_Reg_acc = accuracy_score(predictions, Y_test) * 100
```

```python
Linear_Reg_acc
```

```python
# RANDOM FOREST CLASSIFER
from sklearn.ensemble import RandomForestClassifier

model_random = RandomForestClassifier(n_estimators = 700,
                                     max_features='auto',
                                     oob_score=True,
                                     random_state=1,
                                     n_jobs=1,
                                      min_samples_leaf=1,
                                      min_samples_split=10                                  
                                )
```

```python
model_random.fit(X_train, Y_train)
predictions_random = model_random.predict(X_test)
```

```python
Random_Forest_Acc = accuracy_score(predictions_random, Y_test) * 100
print("Random FOrest acc: {}".format(Random_Forest_Acc))
```

```python
# Support vector machines

from sklearn.svm import SVC, LinearSVC

svc_model = SVC()

svc_model.fit(X_train_scaled, Y_train)
predictions_svc = svc_model.predict(X_test_scaled)
```

```python
SVC_acc = accuracy_score(predictions_svc, Y_test) * 100
print("SVC accuracy: {}".format(SVC_acc))
```

```python
# KNN Class

from sklearn.neighbors import KNeighborsClassifier
n_model  = KNeighborsClassifier(n_neighbors = 4)
n_model.fit(X_train_scaled, Y_train)
```

```python
predictions_knn = n_model.predict(X_test_scaled)
```

```python
accuracy_score(predictions_knn, Y_test) * 100
```

```python
knn_data = {}
for i in range(1,30):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train_scaled, Y_train)
    prediction_data = model.predict(X_test_scaled)
    knn_data[i] = accuracy_score(prediction_data, Y_test) * 100
```

```python
# Naives Bayes

from sklearn.naive_bayes import GaussianNB

model_naive = GaussianNB()

model_naive.fit(X_train, Y_train)

prediction_naive = model_naive.predict(X_test)
```

```python
gaussian_acc = accuracy_score(prediction_naive, Y_test) * 100
print("Gaussian Acc: {}".format(gaussian_acc))
```

```python
# DecisionTree
from sklearn.tree import DecisionTreeClassifier

tree_data = {}
for i in range(2,40):
    tree_model = DecisionTreeClassifier(criterion='gini',
                                    min_samples_split=i,
                              max_features='auto',
                              min_samples_leaf=1)
    tree_model.fit(X_train, Y_train)
    tree_predict = tree_model.predict(X_test)
    tree_data[i]=accuracy_score(tree_predict, Y_test)
```

```python
tree_val = max(tree_data, key=tree_data.get)
print(tree_val)
```
