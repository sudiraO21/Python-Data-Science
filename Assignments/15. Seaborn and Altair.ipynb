````python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
````

````python
df = pd.read_csv('https://raw.githubusercontent.com/vaksakalli/datasets/master/diamonds.csv')
````

````python
df.head()
````

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

````python
fig, ax = plt.subplots(figsize=(13, 13))
palette="ch:r=-.2,d=.3_r"

clarity_order = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="carat", y="price", hue="clarity", size="depth", palette=palette,
                hue_order=clarity_order, sizes=(1, 8), edgecolor=None, data=df, ax=ax)
````

````python
#code here

def hexbin(x, y, color, **kwargs):
    cmap = sns.light_palette(color, as_cmap=True)
    plt.hexbin(x, y, gridsize=15, cmap=cmap, **kwargs)

with sns.axes_style("dark"):
    grid = sns.FacetGrid(df, col="cut", hue="cut", height=4)
grid.map(hexbin, "carat", "price", extent=[0, 3, 20000, 0.5])
````

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price sehingga dihasilkan output sesuai expected.

````python
df = df.sample(n=300, random_state=123)

#code here
sns.pairplot(df, hue='cut', hue_order=['Premium','Ideal','Very Good','Good','Fair'], 
             vars=['carat', 'depth', 'table', 'price'], corner=True, markers=["+","o","s", "D","x"])
````

# Soal 4: Joint Plot

Gunakan data yang sama dengan soal 3 untuk menghasilkan dua visualisasi joint plot antara carat dan price nya sesuai expected output.

````python
#code here
from scipy import stats

# Plot 1
sns.jointplot("carat", "price", data=df, kind='kde', xlim=(-0.3,2.7), ylim=(-2500,20000));

# Plot 2
sns.jointplot("carat", "price", data=df, kind='reg', stat_func=stats.pearsonr, xlim=(0,3), ylim=(0,20000))
````
