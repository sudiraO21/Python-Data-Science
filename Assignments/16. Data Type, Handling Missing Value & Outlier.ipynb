Jawab/Kerjakan pertanyaan/perintah di bawah ini dengan bahasa kalian sendiri:

- Apa itu machine learning?

Machine learning adalah disiplin ilmu yang mencakup perancangan dan pengembangan algoritma yang memungkinkan komputer untuk mengembangkan perilaku yang didasarkan pada data empiris. Sederhananya ini adalah proses komputer untuk belajar dari data (learn from data).
- Apa itu data feature dan data target?

Data feature merupakan data ciri atau unik untuk memprediksi suatu hasil. Hasil inilah yang disebut data target.
- Apa Perbedaan Supervised Learning dan Unsupervised Learning?

Supervised learning adalah sebuah pendekatan dimana sudah terdapat data yang dilatih, dan terdapat variable yang ditargetkan sehingga tujuan dari pendekatan ini adalah mengkelompokan suatu data ke data yang sudah ada, lain halnya dengan unsupervised learning, unsupervised learning tidak memiliki data latih, sehingga dari data yang ada, kita mengelompokan data tersebut menjadi 2 bagian atau 3 bagian dan seterusnya.
- Apa Jenis2 yang ada di dalam Supervised Learning? Jelaskan Perbedaannya!

Supervised Learning dapat dikelompokkan menjadi regresi dan klasifikasi. Ciri dari problem regresi ini adalah kasusnya diminta menebak angka. Jenis outputnya bersifat continuous. Sedangkan dari problem klasifikasi yang diprediksi adalah class, atau jenis. Outputnya bersifat discrete, dan class-nya terbatas.
- Apa perbedaan Hyperparameter dan Parameter?

Parameter nilainya bisa didapatkan secara langsung saat proses training model sedangkan Hyperparameter tidak bisa didapatkan secara langsung pada proses training. Hyperparameter nilainya fix sebelum proses training model.

<h1>Soal 2: Pemahaman Machine Learning</h1>

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian aplikasikan untuk membuat model bagi data (variable x, y) di bawah ini. Kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)
x_newM = x_new[:, np.newaxis]
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
from sklearn.model_selection import train_test_split

x_matriks = x[:, np.newaxis]

X_train, X_test, y_train, y_test = train_test_split(x_matriks, y, random_state=0)

m = DecisionTreeRegressor()

m.fit(X_test,y_test)
```

```python
y_new = m.predict(x_newM)
```

```python
y_train = m.predict(X_train)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(x, y, color = 'b', label = 'Data Asli')
ax.scatter(X_train, y_train, color = 'y', label = 'Prediksi Terhadap Data Training')
ax.plot(x_test, y_new, color = 'g', label = 'Prediksi Terhadap Data Baru')
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
fig.legend()
plt.show()
```
