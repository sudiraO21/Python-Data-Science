<h2>Soal 1: Pivoting and Melting</h2>

Jelaskan itu pivoting dan melting data

jawaban: Pivoting data adalah suatu pengubahan bentuk data dengan memutar data yang terletak di baris menjadi di column sehingga teradi perubahan pada bentuk data dari yang awalnya panjang menjadi lebar. Sedangkan melting data adalah kebalikannya. Melting mengubah suatu data dengan memutar dari yang tadinya berada di posisi column menjadi di posisi row. Melting mengubah data yang tadinya lebar menjadi panjang.

<h2>Soal 4: Reshaping Data 2</h2>

```python
import pandas as pd

data = {
    'Hari' : ['Sabtu', 'Minggu'],
    'Kebun Binatang' : [271, 399],
    'Dufan' : [501, 700],
    'Disney' : [1000, 1001],
    'Bali' : [900, 803]
}
df = pd.DataFrame(data)
df.set_index('Hari')
```

diatas adalah data tentang jumlah pengunjung di berbagai tempat hiburan, ubahlah struktur data tersebut sehingga sesuai dengan hasil yang diharapkan.

```python
pd.melt(df, id_vars = ['Hari'], value_vars =['Kebun Binatang','Dufan','Disney', 'Bali'], 
        var_name ='Tempat Hiburan', value_name ='Pengunjung')
```
