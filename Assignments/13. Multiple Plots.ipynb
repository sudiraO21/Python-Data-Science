```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(ncols=3, figsize=(15,8), sharex=True, sharey=True)

ax[0].title.set_text('Grafik tanx')
ax[1].title.set_text('Grafik sinx')
ax[2].title.set_text('Grafik cosx')

ax[0].plot(x, tan, 'b', label = 'tanX')
ax[1].plot(x, sin, 'r', label = 'sinX')
ax[2].plot(x, cos, 'g', label = 'cosX')

fig.legend(loc='upper center')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
fig, ax = plt.subplots(figsize=(15,10))

ax.title.set_text('Grafik CosX')
ax.plot(x, y,'b')

ins1 = ax.inset_axes([-0.2, 0.9, 0.3 ,0.3])
ins1.title.set_text('Grafik CosX^5')
ins1.plot(x, y5, 'y')
ins1.plot(x, y,'b')

ins2 = ax.inset_axes([0.9, 0.9, 0.3 ,0.3])
ins2.title.set_text('Grafik CosX^4')
ins2.plot(x, y4, 'k')
ins2.plot(x, y,'b')

ins3 = ax.inset_axes([-0.2, -0.1, 0.3 ,0.3])
ins3.title.set_text('Grafik CosX^2')
ins3.plot(x, y2, 'r')
ins3.plot(x, y,'b')

ins4 = ax.inset_axes([0.9, -0.1, 0.3 ,0.3])
ins4.title.set_text('Grafik CosX^3')
ins4.plot(x, y3, 'g')
ins4.plot(x, y,'b')
```
